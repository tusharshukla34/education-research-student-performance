```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\HP\Desktop\Cybrom work\Projects\Education Research Student Performance\Data\student_mat_cleaned.csv")
df.head()
```

```python
# we will check for the numeric column and store in a variable

numeric_df = df.select_dtypes(include='number')
numeric_df.columns
```

```python
# Now i will perform correlation matrix

corr_matrix = numeric_df.corr()
corr_matrix
```

```python
# heat map  
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix,cmap='coolwarm',annot=True,fmt=".2f")
plt.title("Correlation Matrix of Student Performance Variables")
plt.tight_layout()
plt.show()
plt.savefig("Heat map example.jpg")
```
